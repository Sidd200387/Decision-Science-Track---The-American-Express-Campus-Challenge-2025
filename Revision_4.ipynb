```python
import os
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import duckdb
%matplotlib inline
import plotly.express as px 

# os.chdir(r"C:\Users\siddu\Desktop\Decision Science Track\Revision")
print(os.getcwd())

# Upload filepaths
train_data_fp = r"C:\Users\siddu\Desktop\Decision Science Track\Revision\train_df_colsort.parquet"
test_data_fp = r"C:\Users\siddu\Desktop\Decision Science Track\Revision\test_df_colsort.parquet"

# Load files into dataframes
train_data_df = pd.read_parquet(train_data_fp)
test_data_df = pd.read_parquet(test_data_fp)

print(train_data_df.shape)
print(test_data_df.shape)
```

# Create offer age, and offer duration

```python

# --- Create 'offer_age' for TRAIN and TEST ---
# (Your existing logic is perfect)
print("Processing offer_age...")
train_data_df['offer_age'] = (train_data_df['id4'] - train_data_df['id12']).dt.days.clip(lower=0).fillna(999).astype('Int16')
test_data_df['offer_age'] = (test_data_df['id4'] - test_data_df['id12']).dt.days.clip(lower=0).fillna(999).astype('Int16')

# --- Create 'offer_duration' for TRAIN and TEST ---
print("Processing offer_duration...")

# 1. Calculate duration for TRAIN and fill NaNs with sentinel value 999
duration_days_series_train = (train_data_df['id13'] - train_data_df['id12']).dt.days
print(f"Train 'offer_duration' nulls: {duration_days_series_train.isna().sum()}")
train_data_df['offer_duration'] = duration_days_series_train.fillna(999).astype('Int16')

# 2. Calculate duration for TEST and fill NaNs with the same sentinel value 999
duration_days_series_test = (test_data_df['id13'] - test_data_df['id12']).dt.days
print(f"Test 'offer_duration' nulls: {duration_days_series_test.isna().sum()}")
test_data_df['offer_duration'] = duration_days_series_test.fillna(999).astype('Int16')

# 3. Check results
print("\n--- Train Data ---")
print(train_data_df[['offer_age', 'offer_duration']].describe())

print("\n--- Test Data ---")
print(test_data_df[['offer_age', 'offer_duration']].describe())
```

```python
# --- Assuming your merged DataFrames are in memory ---

# Define your output filepaths
train_output_fp = 'train_df_date.parquet'
test_output_fp = 'test_df_date.parquet'

print(f"Saving training data to {train_output_fp}...")
# Use .to_parquet() to save
# index=False is important to avoid saving the pandas index as a separate column
train_data_df.to_parquet(train_output_fp, index=False)

print(f"Saving test data to {test_output_fp}...")
test_data_df.to_parquet(test_output_fp, index=False)


print("Save complete.")
```
